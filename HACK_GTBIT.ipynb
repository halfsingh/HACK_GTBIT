```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelBinarizer,LabelEncoder
```

```python
import glob
import cv2
```

```python
file_card = [cv2.imread(file) for file in glob.glob('dataset/cardboard/*.jpg')]
```

```python
r = 100/file_card[1].shape[1]
```

```python
dim = (100, int(file_card[1].shape[1]*r))
```

```python
def resize_all(to_be_resized):
    resized_X = []
    m = len(to_be_resized)
    for i in range(m):
        resized = cv2.resize(to_be_resized[i], dim, interpolation = cv2.INTER_AREA)
        resized_X.append(resized)
    return resized_X
```

```python
file_card = resize_all(file_card)
```

```python
file_metal = [cv2.imread(file) for file in glob.glob('dataset/metal/*.jpg')]
```

```python
file_metal = resize_all(file_metal)
```

```python
file_paper = [cv2.imread(file) for file in glob.glob('dataset/paper/*.jpg')]
```

```python
file_paper = resize_all(file_paper)
```

```python
file_glass = [cv2.imread(file) for file in glob.glob('dataset/glass/*.jpg')]
```

```python
file_glass = resize_all(file_glass)
```

```python
file_plastic = [cv2.imread(file) for file in glob.glob('dataset/plastic/*.jpg')]
```

```python
file_plastic = resize_all(file_plastic)
```

# Transforming lists of images into arrays

```python

def showimg(file):
    plt.imshow(file[...,[2,1,0]])
    plt.axis('OFF')
    plt.show()
```

```python
for i in range(5):
    showimg(file_card[i])
    showimg(file_metal[i])
    showimg(file_paper[i])
    showimg(file_glass[i])
```

```python
file_card = np.asarray(file_card)
```

```python
file_metal = np.asarray(file_metal)
```

```python
file_glass = np.asarray(file_glass)
```

```python
file_paper = np.asarray(file_paper)
```

```python
file_plastic = np.asarray(file_plastic)
```

```python
file_x = np.vstack((file_card,file_glass,file_metal,file_paper,file_plastic))
```

```python
labels = []
temp = 0
for i in range(200):
    temp = 'cardboard'
    labels.append(temp)


for i in range(200):
    temp = 'glass'
    labels.append(temp)
    

for i in range(200):
    temp = 'metal'    
    labels.append(temp)


for i in range(200):
    temp = 'paper'
    labels.append(temp)
    

for i in range(200):
    temp = 'plastic'
    labels.append(temp)
```

```python
labels = np.asarray(labels)
```

```python
for i in range(0,500,100):
    showimg(file_x[i])
    print(labels[i])
```

```python
from keras.models import Sequential
from keras.layers import Input,Convolution2D,MaxPooling2D,Flatten,Dense,Dropout
from keras.utils import np_utils
```

```python
le = LabelEncoder()
```

```python
labels.shape
```

```python
labels = le.fit_transform(labels)
```

```python
#labels = np_utils.to_categorical(labels)
```

```python
file_x.shape
```

# Building a CNN and training

```python
model = Sequential()
model.add(Convolution2D(32,(3,3),activation = 'relu',input_shape=(100,100,3)))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.25))
model.add(Convolution2D(64,(5,5),activation = 'relu'))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.25))
model.add(Convolution2D(16,(3,3),activation = 'relu'))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(5,activation = 'softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',metrics=["accuracy"],optimizer='adam')
```

```python
model.fit(file_x,labels,batch_size=256,epochs = 20,validation_split=0.20)
```
